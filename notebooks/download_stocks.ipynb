```python
import sys
import platform
import yfinance as yf
import pandas as pd
import time

print("Python:", sys.version)
print("Platform:", platform.platform())
print("yfinance:", yf.__version__)
print("pandas:", pd.__version__)
```

```python
ticker = "GOOGL"

start_time = time.time()
print(f"[{time.strftime('%H:%M:%S')}] Starting download for {ticker}")

df = yf.download(
    tickers=ticker,
    period="1y",
    interval="1d",
    progress=False,
)

elapsed = time.time() - start_time
print(f"[{time.strftime('%H:%M:%S')}] Completed download for {ticker} | rows={len(df)} | elapsed={elapsed:.2f}s")
df.head()
```

```python
tickers = ["AAPL", "MSFT", "GOOGL"]
for t in tickers:
    start_time = time.time()
    print(f"[{time.strftime('%H:%M:%S')}] Starting download for {t}")
    
    df = yf.download(
        tickers=t,
        period="1y",
        interval="1d",
        progress=False,
    )
    
    elapsed = time.time() - start_time
    print(f"[{time.strftime('%H:%M:%S')}] Completed download for {t} | rows={len(df)} | elapsed={elapsed:.2f}s")
```

```python
for t in tickers:
    start_time = time.time()
    print(f"[{time.strftime('%H:%M:%S')}] Fetching {t}")
    
    df = yf.download(
        tickers=t,
        period="1y",
        interval="1d",
        progress=False,
    )
    
    elapsed_fetch = time.time() - start_time
    save_path = f"{t.replace('.', '_')}.csv"
    
    start_save = time.time()
    df.to_csv(save_path)
    elapsed_save = time.time() - start_save
    
    print(f"[{time.strftime('%H:%M:%S')}] Fetched {t} | rows={len(df)} | fetch={elapsed_fetch:.2f}s | save={elapsed_save:.2f}s → {save_path}")

```
