### Extract files from all subfolder

```python
import os, shutil

src = r"D:\Explorer\Series\Sugar.2024.S01.1080p.WEBRip.x265-KONTRAST\Subs"
dst = r"D:\Explorer\Series\Sugar.2024.S01.1080p.WEBRip.x265-KONTRAST\Extracted_SDH"
os.makedirs(dst, exist_ok=True)

for root, _, files in os.walk(src):
    for f in files:
        if "eng,SDH" in f and f.endswith(".srt"):
            episode = os.path.basename(root)   # subdir name (episode)
            new_name = f"{episode}.srt"
            shutil.copy(os.path.join(root, f), os.path.join(dst, new_name))
```

```python
import os, shutil

# Use raw strings for Windows paths
src = r"D:\Explorer\Series\Sugar.2024.S01.1080p.WEBRip.x265-KONTRAST\Subs"
dst = r"D:\Explorer\Series\Sugar.2024.S01.1080p.WEBRip.x265-KONTRAST\Extracted_SDH"
os.makedirs(dst, exist_ok=True)

for root, _, files in os.walk(src):
    for f in files:
        if f.endswith("eng,English [SDH].srt"):   # match pattern
            episode = os.path.basename(root)   # subdir name (episode)
            # If multiple matches exist in the same episode, add a counter
            base_name = f"{episode}"
            new_name = base_name + ".srt"
            counter = 1
            while os.path.exists(os.path.join(dst, new_name)):
                new_name = f"{base_name}_{counter}.srt"
                counter += 1
            shutil.copy(os.path.join(root, f), os.path.join(dst, new_name))
```
