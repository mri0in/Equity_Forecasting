```python
''' Inspect available runs '''

from pathlib import Path
import json

runs_root = Path("c:/Users/Admin/Equity_Forecasting/datalake/runs")

runs = sorted(runs_root.glob("RUN_*"))

for run in runs:
    metrics_path = run / "ensemble" / "metrics.jsonl"
    if not metrics_path.exists():
        continue

    # read last metrics entry
    with open(metrics_path) as f:
        last = list(f)[-1]
        metrics = json.loads(last)

    print(run.name, metrics)
```

```python
run_id = promote_latest_global_signal(
    metric="mean_rmse",
    higher_is_better=False,
    notes="Manual promotion after notebook inspection",
)

print(f"Promoted global signal from run: {run_id}")
```
